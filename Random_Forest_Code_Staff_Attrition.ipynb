If you look at this data, the total staff with attrition is roughly 16% - which means already 16% of a class total whose prediction would be on interest(attrition) is not in the data while 84% of the data does not include any attrition. This is a problem = we have a class imbalance and our prediction is already biased even before we develop the prediction model.

So, how do we address this. One of the ways is to use oversampling. What other ways can we use?... For today, lets use SMOTE from sklearn

```python
#install the imbalanced-learn
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

# Assuming 'final_dataset' contains categorical features represented as strings
# Encode categorical features to numerical values
label_encoder = LabelEncoder()
for column in final_dataset.columns:
    if final_dataset[column].dtype == 'object':
        final_dataset[column] = label_encoder.fit_transform(final_dataset[column])

X = final_dataset.drop(['Attrition'], axis=1)
Y = final_dataset['Attrition']
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.75, stratify=Y)

smote = SMOTE()
X_smote, y_smote = smote.fit_resample(x_train, y_train)
x_train, y_train = X_smote, y_smote
```

Now, let's check if the values are equal

```python
y_train.value_counts()
```

The dataset is balanced and ready for modelling

NOW, LET US BUILD THE PREDICTION MODEL

1. LET US USE A RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=46)
random_forest.fit(x_train,y_train)
```

Hooray! Our model has trained the data(75%), now lets test the remaining 25%

```python
from sklearn.metrics import accuracy_score

prediction = random_forest.predict(x_test)
print(accuracy_score(prediction,y_test))
```

As you can see, our model is not badly off at 85%
